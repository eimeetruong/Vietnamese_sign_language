```python
SEQUENCE_LENGTH = 25
IMAGE_HEIGHT,IMAGE_WIDTH = 220, 340

CLASSES_LIST = ['ban','cochuyengisao','dangkia','motvai','nha','nhinthay','noi','thich','toithichban','xinchao']
```

```python
plt.figure(figsize=(20, 20))

# Get the names of all classes/categories in the dataset
all_classes_names = os.listdir('/content/drive/MyDrive/PBL6_vietnamese/vietnamese_1')

# Generate a list of 20 random values. The values will be between 0-50,
# where 50 is the total number of classes in the dataset.
random_range = random.sample(range(len(all_classes_names)), 10)

# Iterate through all the generated random values
for counter, random_index in enumerate(random_range, 1):

    # Check if the random index is within the range of class names list
    if random_index < len(all_classes_names):
        # Retrieve a class name using the random index
        selected_class_name = all_classes_names[random_index]

        # Retrieve the list of all the video files present in the randomly selected Class Directory.
        video_file_names_list = os.listdir(f'/content/drive/MyDrive/PBL6_vietnamese/vietnamese_1/{selected_class_name}')

        # Check if there are video files in the selected class
        if video_file_names_list:
            # Randomly select a video file from the list retrieved from the randomly selected Class Directory.
            selected_videoFileName = random.choice(video_file_names_list)

            # Initialize a VideoCapture object to read from the video File.
            videoReader = cv2.VideoCapture(f'/content/drive/MyDrive/PBL6_vietnamese/vietnamese_1/{selected_class_name}/{selected_videoFileName}')

            # Read the first frame of the video file
            _, bgr_frame = videoReader.read()

            # Release the VideoCapture object
            videoReader.release()

            # Convert the frame from BGR into RGB format.
            rgb_frame = cv2.cvtColor(bgr_frame, cv2.COLOR_BGR2RGB)

            # Write the class name on the video frame
            cv2.putText(rgb_frame, selected_class_name, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

            # Display the frame
            plt.subplot(5, 4, counter)
            plt.imshow(rgb_frame)
            plt.axis('off')
        else:
            print(f"No video files found in {selected_class_name}")

plt.tight_layout()
plt.show()
```
